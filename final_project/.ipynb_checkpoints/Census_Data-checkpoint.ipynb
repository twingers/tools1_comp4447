This file is for pulling the estimated popluation based upon the 2010 census.  The census data will be used in the COMP4447 final project to compare air pollution to population growth.  


```python
#list of libraries

import numpy as np
import pandas as pd
import re
```

```python
#obtain the files from the US census website
!wget -O Census_CO_2000_2009.xlsx https://www2.census.gov/programs-surveys/popest/tables/2000-2010/intercensal/county/co-est00int-01-08.xls
!wget -O Census_CO_2010_2019.xlsx https://www2.census.gov/programs-surveys/popest/tables/2010-2019/counties/totals/co-est2019-annres-08.xlsx
```

2010 data cleaning

```python
#download to a dataframe remove unneeded title rows
census_df = pd.read_excel("Census_CO_2010_2019.xlsx", skiprows =range(0,3))

#rename first colunm names:
census_df.rename(columns={'Unnamed: 0':'County'}, inplace=True)

#delete unneeded columns:
census_df.drop(["Census", "Estimates Base"], axis =1, inplace = True)


#delete unneeded rows with notes then reset the index 
census_df = census_df.iloc[:-5 , :]
census_df = census_df.iloc[1:, :]

census_df.reset_index(inplace = True) #need to reindex for next for loop to work
census_df.drop(["index"], axis =1, inplace = True)

#delete period in front of county name then remove " County, Colorado"
for i in range(len(census_df)):
    word = re.sub(r'[.]', '', census_df['County'][i])
    pr = word.partition(" ")
    census_df.at[i, 'County'] = pr[0]

census_df.head()
```

2000 data cleaning

```python
#download to a dataframe remove unneeded title rows
census2000_df = pd.read_excel("Census_CO_2000_2009.xlsx", skiprows =range(0,3))
census2000_df.head()

#rename first colunm names:
census2000_df.rename(columns={'Unnamed: 0':'County'}, inplace=True)

#delete unneeded columns:
census2000_df.drop(["Unnamed: 1", "Unnamed: 12", "Unnamed: 13"], axis =1, inplace = True)

#delete unneeded rows with notes then reset the index 
census2000_df = census2000_df.iloc[:-7 , :]
census2000_df = census2000_df.iloc[1:, :]

census2000_df.reset_index(inplace = True) #need to reindex for next for loop to work
census2000_df.drop(["index"], axis =1, inplace = True)

#delete period in front of county name then remove " County, Colorado"
for i in range(len(census2000_df)):
    word = re.sub(r'[.]', '', census2000_df['County'][i])
    pr = word.partition(" ")
    census2000_df.at[i, 'County'] = pr[0]
    
census2000_df.head()
    
```

```python
#merge the two df
census_final = pd.merge(census2000_df, census_df,left_index=True, right_index=True, how='inner') 

#remove the County on the right df
census_final.drop(["County_y"], axis =1, inplace = True)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
census_final.head()
```

```python
#transpose the df and adjust the indexes
#need to transpose since the numbers in the column names cause issues calling the columns

def census_transpose(df):            
    df_T = df.T 
    df_T.rename(columns = df_T.iloc[0], inplace = True)
    df_T.drop(df_T.index[0], inplace = True)
    return df_T

census_final = census_transpose(census_final)
```

```python
#change the dtypes of the columns

census_final = census_final.astype(int)
census_final.head()
```
